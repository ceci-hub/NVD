```python
import pandas as pd
from plotnine import *
import matplotlib.pyplot as plt

plt.style.use('default')

orden = [
    "Very_Excellent", "Excellent", "Very_Good", "Good",
    "Above_Average", "Average", "Below_Average",
    "Fair", "Poor", "Very_Poor"
]


ames = ames.copy()
ames['Overall_Qual'] = pd.Categorical(ames['Overall_Qual'],
                                      categories=orden,
                                      ordered=True)

p = (
    ggplot(ames, aes(x='Overall_Qual'))              
    + geom_bar(fill='grey', color='white', size=1.0)  
    + coord_flip()                                    
    + labs(
        title='Distribución de las casas vendidas según calidad',
        x='Calidad General',                          
        y='Cantidad'
    )
    + theme_minimal()
    + theme(
        figure_size=(10, 6),
        panel_background=element_rect(fill='white', color=None),
        plot_background=element_rect(fill='white', color=None),
        panel_grid_major=element_line(color='#dddddd', size=0.3),
        panel_grid_minor=element_line(color='#eeeeee', size=0.2),
        plot_title=element_text(size=14, weight='bold')
    )
)

p
```

```python
fig, ax = plt.subplots()
fig.suptitle('Distribución de las casas vendidas según calidad') 
sns.countplot( y="Overall_Qual", data=ames, color='grey',
                  order = ["Very_Excellent",
                        "Excellent",
                        "Very_Good",
                        "Good",
                        "Above_Average",
                        "Average",
                        "Below_Average",
                        "Fair",
                        "Poor",
                        "Very_Poor"])
ax.set(yticklabels=["10-Very_Excellent", "9,", "8",
                  "7","6","5-Average","4","3","2",
                   "1-Very_Poor"])
```
